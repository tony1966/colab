```python
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation #新增(Activation)
from tensorflow.keras import optimizers
from tensorflow.keras.utils import plot_model
from tensorflow.keras.utils import to_categorical
import matplotlib.pyplot as plt
import numpy as np
from tensorflow.keras.initializers import Zeros, RandomNormal, glorot_normal, glorot_uniform #新增
```

```python
(X_train, y_train), (X_test, y_test)=mnist.load_data()
```

```python
X_train=X_train.reshape(60000, 784).astype('float32')
X_test=X_test.reshape(10000, 784).astype('float32')
X_train /= 255
X_test /= 255
```

```python
y_train=to_categorical(y_train, 10)
y_test=to_categorical(y_test, 10)
```

```python
b_init=Zeros()       #將b參數初始值設為0
w_init=glorot_uniform()   #將w參數初始值設為標準常態分布取樣
model=Sequential()
model.add(Dense(64, 
        input_dim=784,
        kernel_initializer=w_init,
        bias_initializer=b_init))
model.add(Activation('sigmoid'))
model.add(Dense(10, activation='softmax'))
```

```python
model.compile(loss='mean_squared_error',
       optimizer=optimizers.SGD(learning_rate=0.01),
       metrics='accuracy')
```

```python
model.fit(X_train, y_train, batch_size=128, epochs=200, verbose=1,
     validation_data=(X_test, y_test))
```
